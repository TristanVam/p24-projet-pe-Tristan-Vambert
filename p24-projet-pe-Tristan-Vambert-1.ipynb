```python
# Globalement, comme je l'ai expliqué pour countries et cities, j'ai trouvé assez dur de faire un lien entre les 3 bases de données.
# J'ai néanmoins essayé avec la latitude et la température ce qui donne un résultat satisfaisant, et qui donne de plus l'occasion d'utiliser un merge.
```

```python
avg_temp_city = weather.groupby("city_name")["avg_temp_c"].mean().reset_index()
avg_temp_city.rename(columns={"temperature": "avg_temp_c"}, inplace=True)

target_cities = ["Vienna", "Paris", "Berlin", "Brussels"]
filtered_cities = cities[cities["city_name"].isin(target_cities)]

result = pd.merge(filtered_cities[["city_name", "latitude"]], avg_temp_city, on="city_name")

# Afficher le résultat
print(result)
```

```python
# il n'y a bien entendu aucune relation du type 1/latitude=k*avg_temp_c,
# mais on voit bien que plus on est au nord plus la température moyenne est faible ce qui est cohérent.
```

```python
# On crée un pivot_table pour résumer la température moyenne et la précipitation totale et ce, mois par mois en utilant les timeseries, et les aggfunc sum et min. 
```

```python


weather["date"] = pd.to_datetime(weather["date"])
weather["month"] = weather["date"].dt.month #on fait du mois par mois

pivot = pd.pivot_table(weather,values=["avg_temp_c", "precipitation_mm"],index="city_name",columns="month",aggfunc={"avg_temp_c": "mean", "precipitation_mm": "sum"})

pivot.columns = [f"{stat}_{month}" for stat, month in pivot.columns] #on renomme les colonnes avec f-string
pivot.reset_index(inplace=True)

print(pivot)
```

```python
# Une des caractéristiques du dérèglement du climat est que les évènements dits "exceptionnels" 
#(ici fortes précipitations, fortes chutes de neiges et forts vents) sont de plus en plus élevés et communs.
#J'ai souhaité vérifier cela sur Berlin par exemple en faisant un programme plus élaboré que les précédents.
# C'est de plus l'occasion d'utiliser une 2e fois datetime pour les années cette fois
```

```python
weather["date"] = pd.to_datetime(weather["date"])
weather["year"] = weather["date"].dt.year
berlin_weather = weather[weather["city_name"] == "Berlin"]

# On calcule les maxima année par année pour les différentes mesures
annual_max = berlin_weather.groupby("year").agg(
    max_snowfall=("snow_depth_mm", "max"),
    max_rainfall=("precipitation_mm", "max"),
    max_wind_speed=("peak_wind_gust_kmh", "max")
).reset_index()

plt.plot(annual_max["year"], annual_max["max_snowfall"], label="Max Chute de Neige (cm)", marker='o')
plt.plot(annual_max["year"], annual_max["max_rainfall"], label="Max précipitation (cm)", marker='o')
plt.plot(annual_max["year"], annual_max["max_wind_speed"], label="Max Vitesse du Vent (km/h)", marker='o')

plt.title("Évolution des maxima annuels à Berlin")
plt.xlabel("Année")
plt.ylabel("Valeur maximale")
plt.legend()
plt.grid(True, linestyle='--')
plt.tight_layout()

plt.show()
```
